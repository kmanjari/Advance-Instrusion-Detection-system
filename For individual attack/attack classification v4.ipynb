```python
def bool_attack1(x):
    if(x != "normal"):
        return "attack"
    else:
        return "normal"
```

```python
#divide training and testing data
X_train = x_new[:58630]
y_train = df_output.iloc[:58630]
X_test = x_new[58630:]
y_test = df_output.iloc[58630:]
```

```python
#train your classifier and classify
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
y_score = rf.fit(X_train,y_train)

#predict
y_pred = rf.predict(X_test)
y_pred_prob = rf.predict_proba(X_test)[::,0]
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
C = confusion_matrix(y_test,y_pred)
```

```python
C
```

```python
def perf_measure_class(C, i):
    TP = C[i, i] 
    FP = np.sum(C, axis=0)[i] - C[i, i]  #The corresponding column for class_i - TP
    FN = np.sum(C, axis=1)[i] - C[i, i] # The corresponding row for class_i - TP
    TN = np.sum(C) - TP -FP -FN
    acc = (TP+TN)/(TP+TN+FP+FN)
    fpr = (FP)/(FP+TN)
    dr = (TP)/(TP+FN)
    
    return fpr, acc, dr
```

```python
for i in range(4):
    fpr, acc, dr = perf_measure_class(C, i)
    print("class "+str(i)+": "+"fpr: "+str(fpr)+" acc: "+str(acc)+" dr: "+str(dr))
```
